```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")
student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
#dict的方法
# student1.keys()
# student1.values()
# student1.items()
subjects = list(student1.keys())
stu1_values = list(student1.values())
stu2_values = list(student2.values())
subjects, stu1_values, stu2_values
#subjects, stu1_values  #tuple
```

```python
#標準寫法

# 1.製作figure
figure = plt.figure(figsize=(8, 5),dpi=240,facecolor='#cccccc') #寬8 高5 

# 2.在1.的figure上製作一個axes
axes = fig.add_subplot(1,1,1)
axes.set_title("學生分數")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
# 3.將圖的數據儲存在axes中
axes.plot(subjects,stu1_values,marker='o',linestyle='-.')
# 4.顯示圖
plt.show()
plt.close()
```

```python
#標準寫法

# 1.製作figure
figure = plt.figure(figsize=(8, 5),dpi=240,facecolor='#cccccc') #寬8 高5 
# 2.在1.的figure上製作一個axes
axes = figure.add_subplot(1,1,1)
axes.set_title("學生分數")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
# 3.將圖的數據儲存在axes中
axes.plot(subjects,stu1_values,marker='o',linestyle='-.',label="學生1")
axes.plot(subjects,stu2_values,marker='o',linestyle='-.',label="學生2")
axes.hlines(85,xmin=-1,xmax=6,colors='red',linestyles='--') #劃一條標準線

print(axes.get_xticks()) 
print(axes.get_xticklabels()) #ticks的描述
print(axes.get_yticks())
print(axes.get_yticklabels())
#axes.set_xticklabels(['chinese','english','math','history','discover'])  #也可修改xticklabels
x_locations = axes.get_xticks()
y_locations = axes.get_yticks()
axes.set_xlim(xmin=min(x_locations)-1,xmax=max(x_locations)+1)  #限制X 軸範圍
axes.set_ylim(ymin=min(y_locations)-10,ymax=max(y_locations)+10) #限制y 軸範圍

#axes.annotate("註解",(0,80))  # annotate 可新增點上面的說明
for x in x_locations:
    axes.annotate(str(stu1_values[x]),(x,stu1_values[x]+1))
    axes.annotate(str(stu2_values[x]),(x,stu2_values[x]+1))


axes.legend() #秀圖示label說明
#axes.grid() #劃格線，沒填X，Y都繪畫線
axes.grid(axis='y') #劃格線，指畫Y線
# 4.顯示圖
plt.show()
plt.close()
```
